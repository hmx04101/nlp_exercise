```python
comment_words = ''
stopwords = set(STOPWORDS)
  
# iterate through the csv file
for val in file['Movie'].values:
      
    # typecaste each val to string
    val = str(val)
  
    #split the value
    tokens = val.split()
      
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
      
    comment_words += " ".join(tokens)+" "
    
    
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, 
                      background_color='black', colormap='Set2', 
                      collocations=False, stopwords = STOPWORDS).generate(comment_words)

#plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
def convert_runtime(val):
    """
    Convert the string number value to a float
     - Remove $
     - Remove commas
     - Convert to float type
    """
    new_val = val.replace('min','')
    return int(new_val)
```

```python
converted_RunTime = file['RunTime'].apply(convert_runtime)
```

```python
def convert_Year(val):
    """
    Convert the string number value to a float
     - Remove $
     - Remove commas
     - Convert to float type
    """
    new_val = val.replace('(','').replace(')', '').replace(' Video', '').replace(' TV Special','').replace('III ','').replace('–','').replace(' TV Movie','')
    return int(new_val)
```

```python
converted_Year = file['Year'].apply(convert_Year)
```

```python
def convert_score(val):
    """
    Convert the string number value to a float
     - Remove $
     - Remove commas
     - Convert to float type
    """
    
    return int(val)
```

```python
def convert_votes(val):
    """
    Convert the string number value to a float
     - Remove $
     - Remove commas
     - Convert to float type
    """
    new_val = val.replace(',','')
    return int(new_val)
```

```python
converted_Votes = file['Votes'].apply(convert_votes)
```

<b>Applying weighted Average</b>

```python
v = file['C_IMDB_Score']        # Number of votes for the movie
R = file['C_Votes']             # Vote average of a movie which is a number from 0 to 10
C = file['C_Votes'].mean()      # The mean vote across the whole report 
m = file['C_IMDB_Score'].quantile(0.7)
```

```python
file['Weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
file.head()
```

```python
movie_rankings = file.sort_values('Weighted_average',ascending=False)
movie_rankings[['Movie','C_Votes','C_IMDB_Score', 'movie_imdb_score_rank','Weighted_average']]
```
